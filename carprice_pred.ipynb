```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("/content/data.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data=data.drop(["Engine Fuel Type","Number of Doors","Transmission Type","Market Category"],axis=1)
```

```python
data.head()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=["Engine HP","Engine Cylinders",],inplace=True)
```

```python
data.isnull().sum()
```

```python
data=pd.get_dummies(data,columns=["Driven_Wheels","Vehicle Size","Vehicle Style"],dtype=int)
```

```python
data.head()
```

```python
cols=["Make","Model"]
for col in cols:
  freq=data[col].value_counts(normalize=True)
  data[col]=data[col].map(freq)
```

```python
# from category_encoders import CountEncoder
# encoder=CountEncoder(cols=["Make","Model"])
# data=encoder.fit_transform(data)
```

```python
data.head()
```

```python
y=data["MSRP"]
x=data.drop("MSRP",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)

```

```python
x_train
```

```python
x_train.boxplot(figsize=(15,6))
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
```
